```python
import pandas as pd
import requests
import wbgapi as wb
```

```python
def fetch_all_market_cap(start_year, end_year):
    """
    Fetches market cap data for ALL countries using wbgapi.
    """
    indicator = 'CM.MKT.LCAP.CD'
    
    # wb.data.DataFrame fetches data directly into a pandas DataFrame
    # economy='all' ensures every country/region is included
    # time=range(start, end) is inclusive of the start but exclusive of the end (standard python range)
    try:
        df = wb.data.DataFrame(
            indicator, 
            economy='all', 
            time=range(start_year, end_year + 1),
            labels=True  # Includes country names instead of just ISO codes
        )
        
        # The result is a wide-format dataframe (years as columns)
        # We can reset the index to make it easier to work with
        df = df.reset_index()
        
        # Optional: Filter out rows that are completely empty (no data for any year)
        df = df.dropna(subset=[col for col in df.columns if col.startswith('YR')])
        
        return df
        
    except Exception as e:
        print(f"Error fetching data: {e}")
        return None

```

```python
# --- Run ---
df_all = fetch_all_market_cap(2020, 2025)

# Display the first few rows
if df_all is not None:
    print(f"Fetched data for {len(df_all)} entities.")
    display(df_all.head())
```

```python
#convert df columns to USD billions for better readability
for col in df_all.columns:
    if col.startswith('YR'):
        df_all[col] = df_all[col]/1e9
```
